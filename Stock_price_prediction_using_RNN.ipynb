```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
from sklearn.preprocessing import MinMaxScaler

# Function to prepare data for RNN
def prepare_data(data, time_steps=60):
    X, y = [], []
    for i in range(len(data) - time_steps):
        X.append(data[i:i + time_steps])
        y.append(data[i + time_steps])
    return np.array(X), np.array(y)

# Define the RNN model with multiple layers and dropout
class StockPriceRNN(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size, dropout_prob=0.2):
        super(StockPriceRNN, self).__init__()
        self.rnn = nn.RNN(
            input_size=input_size,
            hidden_size=hidden_size,
            num_layers=num_layers,
            dropout=dropout_prob,  # Dropout applied between layers
            batch_first=True
        )
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        out, _ = self.rnn(x)  # Output shape: [batch_size, time_steps, hidden_size]
        out = self.fc(out[:, -1, :])  # Take the last time step's output
        return out

# Function to train and evaluate the RNN
def fetch_and_predict_stock_price_rnn(ticker, time_steps=60, hidden_size=50, num_layers=3, epochs=30, lr=0.001, dropout_prob=0.2):
    # Fetch data from yfinance
    print(f"Fetching data for {ticker}...")
    data = yf.download(ticker, start="2014-01-01", end="2024-01-01")['Close'].values

    # Scaling the data
    scaler = MinMaxScaler(feature_range=(0, 1))
    data_scaled = scaler.fit_transform(data.reshape(-1, 1))

    # Prepare data for RNN
    X, y = prepare_data(data_scaled, time_steps)
    X_train = torch.tensor(X, dtype=torch.float32)
    y_train = torch.tensor(y, dtype=torch.float32).view(-1, 1)

    # Ensure input is 3D: [batch_size, time_steps, input_size]
    X_train = X_train.reshape(-1, time_steps, 1)  # Input size = 1 for univariate series

    # Initialize the RNN model
    model = StockPriceRNN(
        input_size=1,
        hidden_size=hidden_size,
        num_layers=num_layers,
        output_size=1,
        dropout_prob=dropout_prob
    )
    criterion = nn.MSELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)

    # Train the model
    print(f"Training model for {ticker}...")
    train_losses = []
    for epoch in range(epochs):
        model.train()
        outputs = model(X_train)  # Forward pass
        loss = criterion(outputs, y_train)
        train_losses.append(loss.item())

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        print(f"Epoch [{epoch + 1}/{epochs}], Loss: {loss.item():.4f}")

    # Plot training loss
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, epochs + 1), train_losses, label='Training Loss')
    plt.title('Training Loss Across Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Evaluate the model on the full dataset
    model.eval()
    with torch.no_grad():
        predictions = model(X_train)  # Forward pass on the full dataset
        predictions = scaler.inverse_transform(predictions.numpy())  # Rescale predictions
        y_actual = scaler.inverse_transform(y_train.numpy())  # Rescale actual values

    # Visualization: Full dataset
    plt.figure(figsize=(12, 6))
    plt.plot(y_actual, color='blue', label='Actual Prices')
    plt.plot(predictions, color='red', label='Predicted Prices')
    plt.title(f"{ticker} Stock Price Prediction (Full Dataset)")
    plt.xlabel("Time")
    plt.ylabel("Close Price")
    plt.legend()
    plt.grid(True)
    plt.show()

# Process each stock and visualize predictions
tickers = ['ONGC.NS', 'HDFCBANK.NS', 'ITC.NS', 'ULTRACEMCO.NS', 'SBIN.NS', 'MARUTI.NS']
for ticker in tickers:
    fetch_and_predict_stock_price_rnn(ticker, epochs=30, dropout_prob=0.2)
```
